```python
import pandas as pd

dir = '/opt/ml/'
mask_path = f'{dir}/submissions/submission_mask1_epoch2.csv'
gender_path = f'{dir}/submissions/submission_gender1_epoch2.csv'
age_path = f'{dir}/submissions/submission_age1_epoch3.csv'
```
